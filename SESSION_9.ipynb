## Converting words into tfidf vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer()
```

```python
features=tf.fit_transform(data)
```

```python
features
```

```python
print(features)#17912->id
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
SSD=[]
for k in range(1,10):
    kmeans=KMeans(n_clusters=k,random_state=10)
    kmeans.fit(features)
    SSD.append(kmeans.inertia_)
plt.plot(range(1,10),SSD);
```

```python
#Applying silhouette score
from sklearn.metrics import silhouette_score
SS=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,random_state=10)
    kmeans.fit(features)
    SS.append(silhouette_score(features,kmeans.predict(features)))
plt.plot(range(2,11),SS)
```

# Suppose no of clusters=5

```python
#Building a cluster model
kmeans=KMeans(n_clusters=5,random_state=10)
kmeans.fit(features)
kmeans.labels_
```

## Appending the dataframe

```python
bbc['Cluster']=kmeans.labels_
```

```python
bbc
```

## Dimensionality Reduction and Visualisation

```python
from sklearn.manifold import TSNE
tsne=TSNE(n_components=2,perplexity=30)
```

```python
feature_tsne=tsne.fit_transform(features)
```

```python
print(feature_tsne)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(feature_tsne[:,0],feature_tsne[:,1],c=bbc['Cluster']);
```
